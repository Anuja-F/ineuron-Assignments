**In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets 
```

```python
iris = datasets.load_iris()
```

```python
iris_data = sns.load_dataset("iris")
iris_data.head()
```

```python
X = iris.data
Y= iris.target
```

```python
PCA = decomposition.PCA(n_components=3)
PCA.fit(X)
X = PCA.fit_transform(X)
```

```python
fig = plt.figure(figsize=(8,8))
ax= Axes3D(fig,elev=-150,azim=110)
ax.scatter(X[:,0],X[:,1],X[:,2],c=Y,s=50,alpha=0.9)
ax.set_xlabel('Setosa',fontsize=15)
ax.set_ylabel('Versicolor',fontsize=15)
ax.set_zlabel('Virginica',fontsize=15)
ax.set_title("IRIS data in 3D",fontsize=20)
ax.grid(True)
plt.show() 
```
